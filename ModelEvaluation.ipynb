### Model Validation

How do you know when to stop training? Or what time you know that model is accurate but its not overfitting the data. 
The process defines 3 different data sets and we call them as follows 
1. Training set
2. Validation set
3. Test set.

As you all know by now that we do not want to touch test set until we are 
satisfied with our model prediction. But to get the model prediction we need to train the model optimally. 

1. So what's the point at which we'll stop training?
2. How to know for how many epochs we need to run the loop ?

We devide the data set into 3 models namely Training set, validation set and the test set.

1. During train- The set on which as the name suggests, will train the model on is training set. We define a number for epoch and see how model is behaving. This model will update the weights and biases using backpropagation only for the training set not for validation set

2. Traing set - Update the weights of model based on backpropagation and see how its predicting

3. Validation set - Check how well the model generalizes. It only tells us if the model is working well on the training set. It does not say anything about the weights but only how well its predicting on the validation set. 

4. Finally the test set of data is saved for checking the accuracy after its trained.
